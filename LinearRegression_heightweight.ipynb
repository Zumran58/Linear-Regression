```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#Load dataset

data=pd.read_csv('/home/jovyan/binder/weight-height.csv')
```

```python
data
```

```python
data.shape
```

```python
data.plot(kind='scatter', x='Height', y='Weight')
```

```python
data.plot(kind='box')

plt.show()
```

```python
data.corr() #Correlation coefficients
```

```python
#change to dataframe variable 

height=pd.DataFrame(data['Height'])

weight=pd.DataFrame(data['Weight'])
```

```python
height
```

```python
#build a linear regression model

lm=linear_model.LinearRegression()
model=lm.fit(height, weight)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(height, weight) #evalutate the model
```

```python
#Predict new value of weight

height_new=71

weight_predict=model.predict(height)

weight_predict
```

```python
#Predict monde values

X=([67, 78, 94])
X=pd.DataFrame(X)
Y=model.predict(X)
Y=pd.DataFrame(Y)
df=pd.concat([X,Y], axis=1, keys=['height_new', 'weight_predict'])
df
```

```python
#visualize the result
data.plot(kind='scatter', x='Height', y='Weight')
#plotting the regression line
plt.plot(height, model.predict(height), color='red', linewidth=2)
#plotting the predict value

plt.show()
```
